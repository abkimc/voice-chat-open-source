```python
import torch
import whisper
import numpy as np
from transformers import pipeline
from IPython.display import display, Audio
import textwrap
from ipywebrtc import AudioRecorder, CameraStream
import time
from models import build_model
import os
import numpy as np

try:
    import tensorflow  # required in Colab to avoid protobuf compatibility issues
except ImportError:
    pass

import torch
import pandas as pd
import whisper
import torchaudio
from kokoro import generate
from ipywebrtc import AudioRecorder, CameraStream
from IPython.display import Audio, display
import ipywidgets as widgets


class ConversationalBot:
    def __init__(self):
        # Initialize models
        self.device = 'cuda' if torch.cuda.is_available() else 'cpu'

        # Initialize Whisper
        self.whisper_model = whisper.load_model("base.en")

        # Initialize DeepSeek
        self.text_generator = pipeline(
            "text-generation",
            model="deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B",
            device=self.device
        )

        # Initialize Kokoro
        self.tts_model = build_model('kokoro-v0_19.pth', self.device)
        self.voicepack = torch.load('voices/af.pt', weights_only=True).to(self.device)

        # Initialize recorder
        self.camera = CameraStream(constraints={'audio': True, 'video': False})
        self.recorder = AudioRecorder(stream=self.camera)

        # Conversation history
        self.conversation_history = []

        # File paths
        self.webm_path = 'recording.webm'
        self.wav_path = 'my_recording.wav'
    def split_text_into_chunks(self, text, max_chars=300):
        """Split text into chunks that Kokoro can process."""
        chunks = textwrap.wrap(text, max_chars, break_long_words=False, break_on_hyphens=False)
        return chunks

    def text_to_speech(self, text):
        """Convert text to speech using Kokoro, handling long text."""
        chunks = self.split_text_into_chunks(text)
        full_audio = []

        for chunk in chunks:
            audio, _ = generate(self.tts_model, chunk, self.voicepack, lang='a')
            full_audio.extend(audio)

        return np.array(full_audio)

    def record_audio(self):
        """Record audio from user."""
        print("Recording... Press the record button to start.")
        display(self.recorder)

        # Clear previous recording if it exists
        if hasattr(self.recorder, 'audio'):
            delattr(self.recorder, 'audio')

        # Wait for recording to complete
        while not hasattr(self.recorder, 'audio'):
            time.sleep(0.1)

        # Save recording
        try:
            with open(self.webm_path, 'wb') as f:
                f.write(self.recorder.audio.value)
        except Exception as e:
            print(f"Error saving recording: {e}")
            return False

        # Convert to WAV
        try:
            # Remove existing WAV file if it exists
            if os.path.exists(self.wav_path):
                os.remove(self.wav_path)

            # Convert webm to wav using ffmpeg
            result = subprocess.run([
                'ffmpeg', '-i', self.webm_path,
                '-ac', '1', '-f', 'wav',
                self.wav_path, '-y',
                '-hide_banner', '-loglevel', 'panic'
            ], check=True)

            # Verify the WAV file was created
            if not os.path.exists(self.wav_path):
                print("Error: WAV file was not created")
                return False

            return True

        except subprocess.CalledProcessError as e:
            print(f"Error converting audio: {e}")
            return False

    def transcribe_audio(self):
        """Transcribe recorded audio using Whisper."""
        try:
            if not os.path.exists(self.wav_path):
                raise FileNotFoundError(f"Audio file not found: {self.wav_path}")

            audio = whisper.load_audio(self.wav_path)
            audio = whisper.pad_or_trim(audio)
            mel = whisper.log_mel_spectrogram(audio).to(self.device)

            options = whisper.DecodingOptions(
                language="english",
                task="transcribe",
                without_timestamps=True
            )

            result = self.whisper_model.decode(mel, options)
            return result.text

        except Exception as e:
            print(f"Error transcribing audio: {e}")
            return None

    def generate_response(self, user_input):
        """Generate response using DeepSeek."""
        try:
            # Add user input to conversation history
            self.conversation_history.append({"role": "user", "content": user_input})

            # Generate response
            response = self.text_generator(
                self.conversation_history,
                max_length=200,
                temperature=0.7,
                top_p=0.9,
                do_sample=True
            )[0]['generated_text']

            # Clean up response to get only the new content
            response = response.split(user_input)[-1].strip()

            # Add response to conversation history
            self.conversation_history.append({"role": "assistant", "content": response})
            return response

        except Exception as e:
            print(f"Error generating response: {e}")
            return "I apologize, but I encountered an error generating a response."

    def chat_turn(self):
        """Handle one turn of conversation."""
        # Record audio
        if not self.record_audio():
            print("Failed to record audio. Please try again.")
            return None, None

        # Transcribe audio
        user_text = self.transcribe_audio()
        if user_text is None:
            print("Failed to transcribe audio. Please try again.")
            return None, None

        print(f"You said: {user_text}")

        # Generate and speak response
        response = self.generate_response(user_text)
        print(f"Assistant: {response}")

        try:
            # Convert response to speech
            audio = self.text_to_speech(response)
            display(Audio(data=audio, rate=24000))
        except Exception as e:
            print(f"Error converting response to speech: {e}")

        return user_text, response

    def start_conversation(self):
        """Start the conversation loop."""
        print("Starting conversation. Press Ctrl+C to end.")
        try:
            while True:
                user_text, response = self.chat_turn()
                if user_text is None or response is None:
                    print("There was an error processing your input. Let's try again.")
                print("\n---\n")
        except KeyboardInterrupt:
            print("\nConversation ended.")
        finally:
            # Cleanup temporary files
            for file_path in [self.webm_path, self.wav_path]:
                if os.path.exists(file_path):
                    try:
                        os.remove(file_path)
                    except Exception as e:
                        print(f"Error removing temporary file {file_path}: {e}")
```

```python
import torch
import whisper
import numpy as np
from transformers import pipeline
from IPython.display import display, Audio
import textwrap
from ipywebrtc import AudioRecorder, CameraStream
import time
from models import build_model
import os
import numpy as np

try:
    import tensorflow  # required in Colab to avoid protobuf compatibility issues
except ImportError:
    pass

import torch
import pandas as pd
import whisper
import torchaudio
from kokoro import generate
from ipywebrtc import AudioRecorder, CameraStream
from IPython.display import Audio, display
import ipywidgets as widgets

def setup_recorder():
    """Create and display a new audio recorder."""
    camera = CameraStream(constraints={'audio': True, 'video': False})
    recorder = AudioRecorder(stream=camera)
    display(recorder)
    return recorder

def save_recording(recorder, filename='recording.webm'):
    """Save the recording from the recorder to a file."""
    if not hasattr(recorder, 'audio'):
        print("No recording found. Please record something first.")
        return False

    try:
        with open(filename, 'wb') as f:
            f.write(recorder.audio.value)
        print(f"Recording saved to {filename}")
        return True
    except Exception as e:
        print(f"Error saving recording: {e}")
        return False

def convert_to_wav(webm_file='recording.webm', wav_file='recording.wav'):
    """Convert webm to wav using ffmpeg."""
    try:
        run([
            'ffmpeg', '-i', webm_file,
            '-ac', '1', '-f', 'wav',
            wav_file, '-y',
            '-hide_banner', '-loglevel', 'panic'
        ], check=True)
        print(f"Converted to {wav_file}")
        return True
    except Exception as e:
        print(f"Error converting audio: {e}")
        return False

def transcribe_audio(wav_file='recording.wav', model=None):
    """Transcribe audio file using Whisper."""
    if model is None:
        model = whisper.load_model("base.en")

    try:
        audio = whisper.load_audio(wav_file)
        audio = whisper.pad_or_trim(audio)
        mel = whisper.log_mel_spectrogram(audio).to(model.device)

        options = whisper.DecodingOptions(
            language="english",
            task="transcribe",
            without_timestamps=True
        )

        result = model.decode(mel, options)
        print(f"Transcribed: {result.text}")
        return result.text
    except Exception as e:
        print(f"Error transcribing audio: {e}")
        return None

def get_ai_response(text, generator=None):
    """Get AI response using DeepSeek."""
    if generator is None:
        device = 'cuda' if torch.cuda.is_available() else 'cpu'
        generator = pipeline(
            "text-generation",
            model="deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B",
            device=device
        )

    try:
        response = generator(
            [{"role": "user", "content": text}],
            max_length=200,
            temperature=0.7,
            top_p=0.9,
            do_sample=True
        )[0]['generated_text']

        # Clean up response to get only the new content
        response = response.split(text)[-1].strip()
        print(f"AI Response: {response}")
        return response
    except Exception as e:
        print(f"Error generating response: {e}")
        return None

def text_to_speech(text, tts_model=None, voicepack=None, device='cuda'):
    """Convert text to speech using Kokoro."""
    if tts_model is None:
        tts_model = build_model('kokoro-v0_19.pth', device)
    if voicepack is None:
        voicepack = torch.load('voices/af.pt', weights_only=True).to(device)

    try:
        # Split text into chunks of 300 characters
        chunks = textwrap.wrap(text, 300, break_long_words=False, break_on_hyphens=False)
        full_audio = []

        # Process each chunk
        for chunk in chunks:
            audio, _ = generate(tts_model, chunk, voicepack, lang='a')
            full_audio.extend(audio)

        audio_array = np.array(full_audio)
        display(Audio(data=audio_array, rate=24000))
        return audio_array
    except Exception as e:
        print(f"Error converting text to speech: {e}")
        return None
```
